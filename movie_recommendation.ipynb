```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = {
    "title": [
        "Inception", "Interstellar", "The Matrix", "Avatar", "Tenet",
        "Avengers", "Iron Man", "Thor", "Captain America", "Black Panther",
        "John Wick", "Mad Max", "Die Hard", "Mission Impossible", "Fast and Furious",
        "Titanic", "The Notebook", "La La Land", "Pride and Prejudice", "A Star Is Born",
        "Jurassic Park", "King Kong", "Godzilla", "Transformers", "Pacific Rim",
        "Toy Story", "Finding Nemo", "Up", "Coco", "Inside Out"
    ],
    "genre": [
        "sci fi action thriller", "sci fi space drama", "sci fi action",
        "sci fi adventure fantasy", "sci fi action thriller",

        "action superhero", "action superhero", "action superhero",
        "action superhero", "action superhero",

        "action thriller", "action adventure", "action thriller",
        "action spy thriller", "action racing adventure",

        "romance drama", "romance love drama", "romance musical drama",
        "romance classic drama", "romance music drama",

        "adventure sci fi", "adventure fantasy", "action monster sci fi",
        "action sci fi", "action sci fi",

        "animation family comedy", "animation family adventure",
        "animation family adventure", "animation fantasy family",
        "animation emotion family"
    ]
}
```

```python
df = pd.DataFrame(movies)
df.head()
```

```python
vectorizer = CountVectorizer()
genre_matrix = vectorizer.fit_transform(df["genre"])
```

```python
similarity = cosine_similarity(genre_matrix)
```

```python
def recommend_movie(movie_name):
    if movie_name not in df["title"].values:
        print("Movie not found in database.")
        return

    index = df[df["title"] == movie_name].index[0]
    scores = list(enumerate(similarity[index]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"\nRecommended movies for '{movie_name}':\n")
    for i in scores[1:6]:
        print(df.iloc[i[0]]["title"])
```

```python
movie_name = input("Enter a movie name: ")
recommend_movie(movie_name)
```
